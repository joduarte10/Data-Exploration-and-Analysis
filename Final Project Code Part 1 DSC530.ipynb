```python
# Check for infinite values
infinite_values = df[['Pressure(in)', 'Humidity(%)', 'Wind_Chill(F)', 'Temperature(F)', 'Visibility(mi)']].replace([float('inf'), float('-inf')], pd.NA).isna().sum()
print(infinite_values)
```

```python
# Replace infinite values with NaN and then handle them
df.replace([float('inf'), float('-inf')], pd.NA, inplace=True)

# Drop rows with NaN values (which include previously infinite values)
df_clean = df.dropna(subset=['Pressure(in)', 'Humidity(%)', 'Wind_Chill(F)', 'Temperature(F)', 'Visibility(mi)'])
```

```python
import statsmodels.api as sm

# Prepare the features and target variable
X = df_clean[['Pressure(in)', 'Humidity(%)', 'Wind_Chill(F)', 'Temperature(F)']]
y = df_clean['Visibility(mi)']

# Add a constant to the model (intercept)
X = sm.add_constant(X)

# Fit the model
model = sm.OLS(y, X).fit()

# Print the summary of the model
print(model.summary())
```

The model explains a small portion (17.9%) of the variance in Visibility(mi), indicating that other factors not included in the model might be influencing visibility.
All predictors are statistically significant, but the model has some potential issues, such as non-normality of residuals and high multicollinearity, which might affect the reliability of the results.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('US_Accidents_March23.csv')

# Data cleaning for Start_Time and End_Time
df['Start_Time'] = df['Start_Time'].str.replace(r'\.\d+$', '', regex=True)
df['Start_Time'] = pd.to_datetime(df['Start_Time'])
df['End_Time'] = df['End_Time'].str.replace(r'\.\d+$', '', regex=True)
df['End_Time'] = pd.to_datetime(df['End_Time'])
df['Duration'] = df['End_Time'] - df['Start_Time']
df['Duration_Hours'] = df['Duration'].dt.total_seconds() / 3600

# Add Day of Week
df['Day_Name'] = df['Start_Time'].dt.day_name()

# Filter Weekdays and Weekends
weekdays = df[~df['Day_Name'].isin(['Saturday', 'Sunday'])]
weekends = df[df['Day_Name'].isin(['Saturday', 'Sunday'])]

# PMF Function
def pmf(data):
    counts = data.value_counts().sort_index()
    probs = counts / len(data)
    return probs

# PMFs
weekday_pmf = pmf(weekdays['Hour'])
weekend_pmf = pmf(weekends['Hour'])

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(weekday_pmf.index, weekday_pmf.values, label='Weekdays')
plt.plot(weekend_pmf.index, weekend_pmf.values, label='Weekends')
plt.title('PMF of Accident Hour (Weekdays vs. Weekends)')
plt.xlabel('Hour of Day')
plt.ylabel('Probability')
plt.legend()
plt.show()
```
